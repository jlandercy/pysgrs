```python
import pysgrs
import warnings
warnings.simplefilter("ignore")
```

```python
[c for c in dir(pysgrs) if c.endswith("Cypher") and not c.startswith("Generic")]
```

```python
s = "A successful man is one who can lay a firm foundation with the bricks others have thrown at him."
```

```python
C = pysgrs.CaesarCypher()
C.cypher(s)
```

```python
C.to_graph()
```

```python
C = pysgrs.VigenereCypher(key="David Brinkley".upper().replace(" ", ""))
C.cypher(s)
```
