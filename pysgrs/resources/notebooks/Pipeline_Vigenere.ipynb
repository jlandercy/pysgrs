```python
import pysgrs
```

```python
s = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

```python
V1 = pysgrs.VigenereCypher("BOB")
V2 = pysgrs.VigenereCypher("BOBETTE")
```

```python
P = pysgrs.PipelineCypher([V1, V2])
```

```python
print(P)
```

```python
a = V1.key*V2.keysize
b = V2.key*V1.keysize
```

```python
V3 = pysgrs.VigenereCypher(key=a)
k = V3.cypher(b)
```

```python
k
```

```python
len(k)
```

```python
V4 = pysgrs.VigenereCypher(key=k)
```

```python
P.cypher(s)
```
