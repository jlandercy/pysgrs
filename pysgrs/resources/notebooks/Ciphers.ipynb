```python
import pysgrs
import warnings
warnings.simplefilter("ignore")
pysgrs.settings.logger.setLevel(40)
```

```python
[c for c in dir(pysgrs) if c.endswith("Cipher") and not c.startswith("Generic")]
```

```python
s = "A successful man is one who can lay a firm foundation with the bricks others have thrown at him."
```

```python
C = pysgrs.CaesarCipher()
C.encipher(s)
```

```python
C.to_graph()
```

```python
C = pysgrs.VigenereCipher(key="David Brinkley".upper().replace(" ", ""))
C.encipher(s)
```

```python
C.to_graph()
```

```python
A = pysgrs.IntegerAlphabet("ABCDEFG")
print(A)
```

```python
C = pysgrs.CaesarCipher(alphabet=A)
print(C)
```

```python
C.to_graph()
```

```python
C.encipher("Aba Fegha")
```
